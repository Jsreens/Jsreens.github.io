```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# GOT CSV file
# https://www.kaggle.com/mylesoneill/game-of-thrones
battleData = pd.read_csv("GOTbattles.csv")
battleData
```

```python
kingWins = battleData.sort_values("name",ascending=False)[["attacker_king","attacker_outcome"]]
```

```python
def getWinLoss(attackerData):
    
    winCount = 0
    lossCount = 0

    for i in attackerData:
        if i == "win":
            winCount += 1
        elif i == "loss":
            lossCount += 1
    return winCount, lossCount
            

```

```python
getWinLoss(battleData["attacker_outcome"])
```

```python
winLossData = getWinLoss(battleData["attacker_outcome"])
x = np.array(["Win","Loss"])
y = ([winLossData[0],winLossData[1]])

plt.title("Wins and Losses of Attacking Kings")
plt.ylabel("Number of Attacks by Kings")
plt.bar(x,y)
plt.show()
```
